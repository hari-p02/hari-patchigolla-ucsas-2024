```python
import pandas as pd
from tensorflow.keras.models import load_model
from tensorflow.keras.models import Model
from sklearn.preprocessing import MinMaxScaler
import numpy as np

pd.set_option('display.max_columns', 500)
```

```python
model = load_model("/content/soccer_player_nn.keras")
```

```python
embedding_model = Model(inputs=model.input, outputs=model.layers[-2].output)
embedding_model.save('embedding_model.h5')
```

```python
df = pd.read_csv("/content/embedding_pre_data.csv")
df.head()
```

```python
scaler = MinMaxScaler()
all_trans = scaler.fit_transform(df.iloc[:, 4:])
```

```python
embds = embedding_model.predict(all_trans)
```

```python
np.save("embds.npy", embds)
```

```python
with open('players_embds.txt', 'w') as file:
    for i in range(embds.shape[0]):
      name = df.loc[i]['short_name']
      lst = embds[i].tolist()
      lststr = " ".join([str(x) for x in lst])
      file.write(f"{name} : " + lststr + '\n')
```
